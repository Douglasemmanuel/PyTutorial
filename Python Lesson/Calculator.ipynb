```python
# Program to create a Simple calculator
# This function adds two numbers
def add(a,b):
    return a + b

# This function substract two numbers
def substract(a,b):
    return a - b

# This function multiply two numbers
def multiply(a,b):
    return a * b

# This function divide two numbers
def divide(a,b):
    return a/b

# This function squares two numbers
def Square(a):
    return a *a

print("Select an Operation")
print("1.Add")
print("2.Substract")
print("3.Multiply")
print("4.Division")
print("5.Square")

while True:
    # Accept input from the User
    choice = input("Enter choice (1/2/3/4/5)")
    # To check if the user picked one of the four option
    if choice in ("1","2","3","4","5"):
        num1 = int(input("Enter your first number:"))
        num2 = int(input("Enter your second number:"))

        if choice == "1":
            print(num1,"+",num2,"=",add(num1,num2))

        elif choice == "2":
            print(num1,"-",num2,"=",substract(num1,num2))

        elif choice == "3":
            print(num1,"*",num2,"=",multiply(num1,num2))
        
        elif choice == "4":
            print(num1,"/",num2, "=",divide(num1,num2))

        elif choice == "5":
            print(num1,"**",num1,"=" ,Square(num1))
        
        # To check if the User wants to perform another Calculation
        # Break the While loop if it is no
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation == "no":
          break
    else:
        print("Invalid Input")
        


    
```
